```python
import pandas as pd
import sqlite3
```

```python
# Make sure you are in the right working directory.
```

```python
## Create connection to testdb.db using sqlite3 & provide db file name.
con = sqlite3.connect("testdb.db")
cur = con.cursor()

## Inspected db file to see what tables exist.
cur.execute("""SELECT * FROM sqlite_master;""")
results = cur.fetchall()
results
```

```python
people = pd.read_sql_query("""select * from people""", con)
people.head(5)

```

```python
## Variables to store names of visitors

first_name = people['first_name']
last_name = people['last_name']
```

```python
visitors = pd.read_sql_query("""select personid, count(personid) as total_visits
               from visits
               group by personid """, con)

visitors

## By inspecting the table, personId[0] needs to be removed, because
## There is no one with that Id in the people table.
```

```python
## personId[0] removed
## first_name and last_name columns added to DataFrame
## total_visits column sorted in descending order
## Table saved as an csv file to be used for Django rest API

visitors.drop(0,inplace=True)
visitors.insert(1, 'first_name', first_name)
visitors.insert(2, 'last_name', last_name)
visitors.sort_values(by="total_visits", ascending=False, inplace=True)
visitors.head(10).to_csv('FrequentBrowsers.csv', index=False)

```
